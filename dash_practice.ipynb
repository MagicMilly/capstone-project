```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
bechdel_df = pd.read_csv('my_data/updated_bechdel_3.csv')
bechdel_df.head()
```

```python
bechdel_df.rename(index=str, columns={'director_fnbt': 'director_gender', 'writer_fnbt': 'writer_gender',
                                     'exec_fnbt': 'exec_gender', 'producer_fnbt': 'producer_gender',
                                     'overall_fnbt': 'overall_gender', 'new_budgets': 'budget',
                                     'new_revenues': 'revenue'}, inplace=True)
```

```python
bechdel_df.to_csv('my_data/updated_bechdel_4.csv', index=False)
```

```python
bechdel_with_budget_and_revenue = bechdel_df.loc[(bechdel_df.budget != 0) & (bechdel_df.revenue != 0)]
bechdel_with_budget_and_revenue.head()
```
